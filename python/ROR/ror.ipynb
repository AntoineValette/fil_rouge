```python
text = """education; funder
facility
facility; funder
education
company; funder
company
healthcare
healthcare; funder
nonprofit; funder
government
nonprofit
other
government; funder
other; funder
archive
archive; funder
company; education
facility; government
facility; government; funder
archive; government
government; healthcare
funder; other
funder; government
funder; nonprofit
funder; healthcare
education; government
funder
facility; nonprofit
company; government
funder; government; healthcare
facility; healthcare
education; facility
facility; funder; government
archive; funder; nonprofit
education; healthcare
company; facility
archive; facility; funder
healthcare; nonprofit
education; funder; healthcare"""

# Utilisation d'un set pour stocker les mots uniques
unique_words = set()

# Parcourir chaque ligne
for line in text.splitlines():
    # Séparer chaque ligne par le délimiteur ';'
    for word in line.split(";"):
        # Supprimer les espaces superflus
        word = word.strip()
        if word:
            unique_words.add(word)

# Affichage du résultat
print(unique_words)
```
